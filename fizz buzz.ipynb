```python
#fizz buzz game
new_list = []
i = 0
def number(x):
    for i in range(1,x+1):
        
        if i % 5 == 0 and i % 3 == 0:
            new_list.append("fizz buzz")
        elif i%3 == 0 :
            new_list.append("fizz")
        elif i%5 == 0 :
            new_list.append("buzz")
        else :
            new_list.append(i)
            i =i+1
number(100)
for x in new_list:
    print(x)
   
```
